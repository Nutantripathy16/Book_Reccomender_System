# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

Now the popular df will be given by 

```python

popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"]>=250].sort_values("avg_rating",ascending=False)
```

```python
popular_df
```

merge in python acts as join in sql

to get the name of book author, image related to it, year of publication and other details, we will join popular_df with book df and our common column will be Book title

```python
popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","num_ratings","avg_rating"]]
```

# Collaborative Filtering Based Recommender System

Here we will also consider, how many people are serious readors and we will consider the feedback of those serious readors to reccomend a book to another user. 

Meanwhile we also need to consider which books are popular, how many ratings they are getting 

```python
ratings_with_name
```

```python
#Here we are checking the number of users who have given more than 200 ratings
users = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200

serious_users = users[users].index
```

```python
filtered_users = ratings_with_name[ratings_with_name["User-ID"].isin(serious_users)]
filtered_users
```

So 474007 ratings are given by just these top 200 serious users

```python
famous_books = filtered_users.groupby("Book-Title").count()["Book-Rating"] > 50
famous_books
```

```python
famous_books = famous_books[famous_books].index
```

```python
famous_books
```

```python
final_ratings = filtered_users[filtered_users["Book-Title"].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

in above pivot table, each book can be seen as a vector in a 679 dimensional space. Now we need to find how these vectors are similar to each other by calculating Euclidien distance between them. 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

Similarity score is signifying the similarity between each book eith each of the book

```python
print(list(enumerate(similarity_score[0])))
```
